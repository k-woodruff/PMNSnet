```python
import torch
import numpy as np
import matplotlib.pyplot as plt

def Pendulum(t, b, kappa, A0=1, m=1, delta0=0):
    
    omega = np.sqrt(kappa/m)*np.sqrt(1-b**2/(4*m*kappa))
    Pendulum_position = A0*torch.exp(-b*t/(2*m))*torch.cos(omega*t + delta0)
    return Pendulum_position

plt.figure(figsize=(7,6))

kappa = 5 
b = 0.5
times = torch.linspace(0,5,50)
plt.scatter(times,Pendulum(times,b,kappa),color='r')

times = torch.linspace(0,10,500)
plt.plot(times,Pendulum(times,b,kappa),color='k')

plt.xticks(fontsize=16)
plt.yticks(fontsize=16)
plt.xlabel("time [s]",fontsize=22)
plt.ylabel("position [m]",fontsize=22)
plt.grid()
plt.show()
```
